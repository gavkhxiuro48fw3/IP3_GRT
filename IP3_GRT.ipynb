```python
data=pd.read_csv("car-sales.csv")
```

```python
data
```

```python
data.describe()
```

5.

```python
data["Price"] = data["Price"].str.replace("$", "").str.replace(",", "").astype(int)
data.head(10)
```

6.

```python
data['Total Sales']=data['Price'].cumsum()
data
```

7.

```python
data['Sale_Date']=pd.date_range(start='2024-02-22',
                                end=None,
                                periods=10,
                                freq='D',
                                tz=None,
                                normalize=False,
                                name=None)
data
```

```python
periods=len('car_sales')
periods
```

8.

```python
fig=plt.figure(figsize=(10,6))

ax5= fig.add_axes([0,0,1,1])

ax5.plot(data['Price'],
        data['Total Sales'],
        color="blue")
plt.title("Price vs Total Sales Line Plot")
plt.xlabel('Price')
plt.ylabel('Total Sales')
fig.show()
```

9.

```python
plt.scatter(data['Odometer (KM)'],
            data['Price'])
plt.title("Odometer vs Price Scatter Plot")
plt.xlabel("Odometer (KM)")
plt.ylabel("Price")
plt.show()
```

10.

```python
plt.bar(data['Make'],
        data['Odometer (KM)'],
        color="blue")
plt.xlabel('Make')
plt.ylabel("Odometer")
plt.title("Make vs Odometer Bar Plot")
plt.show()
```

11.

```python
plt.hist(data['Make'],
        color="blue")
plt.xlabel("Make")
plt.ylabel("Frequency")
plt.title("Histogram of car make")
plt.show()
```

12.

```python
fig=plt.figure(figsize=(15,7))
ax1=fig.add_subplot(221)
ax1.set_title("Price vs Total Sales lineplot")
ax1.plot(data['Price'],
        data['Total Sales'])
ax2=fig.add_subplot(222)
ax2.set_title("Odometer vs Price scatterplot")
ax2.scatter(data['Odometer (KM)'],
        data['Price'])
ax3=fig.add_subplot(223)
ax3.set_title("Make vs Odometer bar plot")
ax3.bar(data['Make'],
        data['Odometer (KM)'],)
ax4=fig.add_subplot(224)
ax4.set_title("Histogram of Car Make")
ax4.hist(data['Make'])

plt.show()
```

13.

```python
sns.boxplot(x='Colour',
           y='Total Sales',
           data=data)
plt.show()
```

14.

```python
sns.lmplot(x='Odometer (KM)',
           y='Total Sales',
           hue="Make",
           ci=False,
           data=data)
plt.show()
```

15.

```python
sns.pairplot(data[['Odometer (KM)',
                       'Doors',
                      'Price',
                      'Total Sales']])
plt.show()
```

16.

```python
corr=data[['Total Sales',
               'Price',
               'Odometer (KM)']].corr()
print(corr)
sns.heatmap(corr, annot=True)
plt.show()
```

There is a strong negative correlation between price and Odometer reading, a small positive correlation between price and  total sales, and a medium negative correlation between total sales and odometer reading.
